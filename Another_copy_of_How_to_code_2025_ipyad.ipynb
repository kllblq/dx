### Conditional Statements: `if`, `elif`, and `else`

Conditional statements allow you to control the flow of your program. They execute a block of code only if a specified condition is true.

*   **`if` statement:** Executes a block of code if the condition is true.
*   **`elif` statement:** (short for "else if") Checks another condition if the previous `if` or `elif` conditions were false.
*   **`else` statement:** Executes a block of code if none of the `if` or `elif` conditions are true.

Here's a basic example:

```python
# Example using if, elif, and else

temperature = 25

if temperature > 30:
  print("It's a hot day!")
elif temperature > 20:
  print("It's a warm day.")
else:
  print("It's a cool day.")
```

In this example:

1.  The `if` statement checks if `temperature` is greater than 30. If it is, the first message is printed.
2.  If the `if` condition is false, the `elif` statement checks if `temperature` is greater than 20. If it is, the second message is printed.
3.  If both the `if` and `elif` conditions are false, the `else` statement is executed, and the third message is printed.

You can have multiple `elif` statements between an `if` and an `else`. The `else` statement is optional.

Here's another example with more conditions:

```python
score = 85

if score >= 90:
  print("Excellent!")
elif score >= 80:
  print("Very Good!")
elif score >= 70:
  print("Good.")
else:
  print("Needs Improvement.")
```

### Indexing in Python

Indexing allows you to access individual characters in a string or items in a list (and other sequence types) by their position.

*   **Zero-based Indexing:** The first element is at index 0.
*   **Positive Indices:** Start from the beginning (0, 1, 2, ...).
*   **Negative Indices:** Start from the end (-1 for the last element, -2 for the second to last, ...).

```python
# Indexing strings
my_string = "Python"

print("First character:", my_string[0])      # Output: P
print("Third character:", my_string[2])      # Output: t
print("Last character:", my_string[-1])       # Output: n
print("Second to last character:", my_string[-2]) # Output: o
```

```python
# Indexing lists
my_list = ["apple", "banana", "cherry", "date"]

print("First item:", my_list[0])        # Output: apple
print("Third item:", my_list[2])        # Output: cherry
print("Last item:", my_list[-1])         # Output: date
print("Second to last item:", my_list[-2]) # Output: cherry
```

### Looping in Python

Looping is a fundamental programming concept that allows you to execute a block of code repeatedly. Python provides two main types of loops: `for` loops and `while` loops.

#### `for` loops

A `for` loop is used to iterate over a sequence (like a list, tuple, string, or range) or other iterable objects. The code block inside the loop is executed once for each item in the sequence.

Here's the basic syntax:

```python
# Example of a for loop iterating over a list
fruits = ["apple", "banana", "cherry"]
for fruit in fruits:
  print(fruit)

# Example of a for loop iterating over a string
for letter in "Python":
  print(letter)

# Example of a for loop using range()
# range(5) generates numbers from 0 up to (but not including) 5
for i in range(5):
  print(i)

# range(start, stop)
for i in range(2, 7):
  print(i)

# range(start, stop, step)
for i in range(0, 10, 2):
  print(i)
```

#### `while` loops

A `while` loop repeatedly executes a block of code as long as a given condition is true.

Here's the basic syntax:

```python
# Example of a while loop
count = 0
while count < 5:
  print(count)
  count += 1 # Increment count to eventually make the condition false

# Example of a while loop with a break statement
# The loop will stop when i equals 3
i = 0
while i < 6:
  print(i)
  if i == 3:
    break
  i += 1

# Example of a while loop with a continue statement
# The continue statement skips the rest of the code in the current iteration
# and moves to the next iteration
j = 0
while j < 5:
  j += 1
  if j == 3:
    continue
  print(j)
```

### Functions in Python

Functions are defined using the `def` keyword, followed by the function name, parentheses `()`, and a colon `:`. The code block within the function is indented.

Here's the basic syntax for defining a function:

```python
def my_function():
  print("Hello from a function!")
```

To execute the code inside a function, you need to call the function by its name followed by parentheses:

```python
my_function() # Calling the function
```

#### Functions with Parameters

Functions can accept arguments (also called parameters) that are passed into the function when it is called. Parameters are specified within the parentheses in the function definition.

```python
def greet(name):
  print(f"Hello, {name}!")

greet("Alice") # Calling the function with an argument
greet("Bob")
```

#### Functions with Return Values

Functions can also return values using the `return` keyword. This allows the function to compute a value and send it back to the part of the code that called the function.

```python
def add_numbers(a, b):
  result = a + b
  return result

sum_result = add_numbers(5, 3) # Calling the function and storing the returned value
print(sum_result)

print(add_numbers(10, 20)) # Calling the function and printing the returned value directly
```

### Slicing in Python

Slicing allows you to extract a portion of a sequence (like strings, lists, or tuples) by specifying a start and end index.

The basic syntax for slicing is `sequence[start:stop:step]`:

*   **`start`**: The index of the first element to include in the slice (inclusive). If omitted, it defaults to the beginning of the sequence.
*   **`stop`**: The index of the element *after* the last element to include in the slice (exclusive). If omitted, it defaults to the end of the sequence.
*   **`step`**: The step size between elements. If omitted, it defaults to 1.

```python
# Slicing strings
my_string = "Programming"

print("Original string:", my_string)
print("Slice from index 3 to 7 (exclusive):", my_string[3:7]) # Output: gram
print("Slice from the beginning to index 5 (exclusive):", my_string[:5]) # Output: Progr
print("Slice from index 6 to the end:", my_string[6:])    # Output: mming
print("Slice with a step of 2:", my_string[::2])       # Output: Prgamn
print("Reverse the string:", my_string[::-1])         # Output: gnimmargorP

# Slicing lists
my_list = [10, 20, 30, 40, 50, 60, 70]

print("\nOriginal list:", my_list)
print("Slice from index 2 to 5 (exclusive):", my_list[2:5])   # Output: [30, 40, 50]
print("Slice from the beginning to index 4 (exclusive):", my_list[:4]) # Output: [10, 20, 30, 40]
print("Slice from index 3 to the end:", my_list[3:])      # Output: [40, 50, 60, 70]
print("Slice with a step of 2:", my_list[::2])         # Output: [10, 30, 50, 70]
print("Reverse the list:", my_list[::-1])           # Output: [70, 60, 50, 40, 30, 20, 10]
```

```python
import yfinance as yf

# Get ticker data for Bitcoin
btc_ticker = yf.Ticker("BTC-USD")

# Get all historical data
historical_data = btc_ticker.history(period="max")

print("Historical data for BTC:")
print(historical_data)
```

```python
# Access the closing price of the first row (index 0)
first_day_close = historical_data['Close'][0]
print(f"The closing price on the first day is: {first_day_close}")

# You can also use .iloc[] for integer-location based indexing
# This gets the value at the first row (index 0) and the 'Close' column (which is the 3rd column, index 3)
first_day_close_iloc = historical_data.iloc[0, historical_data.columns.get_loc('Close')]
print(f"Using iloc, the closing price on the first day is: {first_day_close_iloc}")

# Or, if you know the date (index of the row), you can use .loc[]
# Replace 'YYYY-MM-DD' with the actual date you want to access
# For example, to get the closing price on '2014-09-17':
close_on_specific_date = historical_data.loc['2014-09-17 00:00:00+00:00', 'Close']
print(f"The closing price on 2014-09-17 is: {close_on_specific_date}")
```

```python
#how to access a value out of one column
#. build a new dataframe based off a condition
df_40k = historical_data.loc[historical_data['Close']!=40000]
print(df_40k)
```

```python
# Access the closing price of the first row (index 0) using column name and integer index
first_day_close = historical_data['Close'][0]
print(f"The closing price on the first day is: {first_day_close}")

# Using .iloc[] for integer-location based indexing
# This accesses the value at the first row (index 0) and the 'Close' column (which is at integer index 3)
first_day_close_iloc = historical_data.iloc[0, 3]
print(f"Using iloc, the closing price on the first day is: {first_day_close_iloc}")

# Using .loc[] for label-based indexing (using the date as the index)
# Replace '2014-09-17 00:00:00+00:00' with the actual date you want to access
close_on_specific_date = historical_data.loc['2014-09-17 00:00:00+00:00', 'Close']
print(f"The closing price on 2014-09-17 is: {close_on_specific_date}")

# To get the value from a new DataFrame created by a condition, like df_40k
# You can access a specific row and column from df_40k in the same ways:
# For example, to get the 'Open' price of the first row in df_40k
if not df_40k.empty: # Check if df_40k is not empty before trying to access elements
    first_row_df40k_open = df_40k.iloc[0]['Open']
    print(f"\nThe Open price of the first row in df_40k is: {first_row_df40k_open}")
else:
    print("\ndf_40k is empty, no specific value to access.")
```
